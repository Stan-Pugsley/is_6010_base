# Create a Regression Model


*   Independent (input) variables: month
*   Dependent (target) variable: total_amount



This first step will summarize the data by month

```python
df2 = df[["month","year","total_amount"]].groupby(["month","year"]).sum("total_amount").reset_index()
df2.head()
```

This step will assign variables to X and Y

```python
x = df2[["month","year"]]
y = df2["total_amount"]
model = LinearRegression().fit(x, y)
```

This step will print the results

```python
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

Calculate the fit (R-squared) of the forecast.  

Remember that 
- 1.0 is a perfect fit
- 0 is a poor fit, no better than using a simple average

```python
print('R-squared:', model.score(x, y))
```
